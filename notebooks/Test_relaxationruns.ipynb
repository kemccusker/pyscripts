```python
import cccmaplots as cplt
import cccmautils as cutl
import cccmaNC as cnc
import constants as con
import cccmacmaps as ccm
import numpy.ma as ma
import pandas as pd
%matplotlib inline

printtofile=False

model = 'CanESM2'
bp=con.get_basepath()
basepath=bp['basepath'] + model + '/'; subdir=bp['subdir']

#casenamec='' # don't have that yet
casenamep='kel11'
timeper='2921-2930'
timesel = '2922-01-01,2930-12-31' # skip first year??


field = 'sicn'
ncfield=field.upper()

seasons = ('SON','DJF','MAM','JJA')

#fnamec = basepath + casenamec + subdir + casenamec + '_' + field + '_' + timeper + '_ts.nc'
fnamep = basepath + casenamep + subdir + casenamep + '_' + field + '_' + timeper + '_ts.nc'
#fnamep2x = basepath + casenamep2x + subdir + casenamep2x + '_' + field + '_' + timeper + '_ts.nc' #@@ process

time = cnc.getNCvar(fnamep,'time',timesel=timesel)
ntime = len(time)

lat = cnc.getNCvar(fnamep,'lat')
lon = cnc.getNCvar(fnamep,'lon')

fldp = cnc.getNCvar(fnamep,ncfield)

#fldp2x = cnc.getNCvar(fnamep2x,ncfield)
```
