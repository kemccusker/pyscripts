```python
# calculate correlations between AO index and snow cover.
# Start by using Z500 averaged over polar cap.

# From Furtado et al. 2015
# Model atmospheric variables analyzed are identical to
# those from observations. Snow cover from the model out-
# put is provided as the monthly-mean fractional area (0–1)
# of a grid cell covered in snow. To get an areal extent of
# snow cover, the fractional area is multiplied by the size
# of the grid box and then summed over the area from 0 ◦ to
# 170°W, 20°N to 75°N, which coincides roughly with the
# same area used for the Rutgers Global Snow Lab Eurasian
# SCE index


import loadLE as le
import cccmautils as cutl

%matplotlib inline


field = 'sce'; ncfield='sce'; comp='LImon'; conv=1; region='eurasiasnow'
fieldz = 'zg50000.00'; ncfieldz='zg'; compz='Amon'; convz=1; regionz='gt65n'

timesel1 = '1950-01-01,1980-12-31' # 31-years
timesel2 = '1979-01-01,2020-12-31'; timestr2 = '1979-2020' # 42 years
#timesel2 = '1979-01-01,2012-12-31'; timestr2 = '1979-2012' # 33 years

mo=10 # october for SCE
seaz='DJF'

local = True
```

```python
def load_field(fdict,casename,timesel,seas=None,mo=0,ftype='fullts',orig=None,calc=None,conv=1,local=False,verb=False):
    
    """ TAKEN FROM load_field() in canesm_le_composite.py/pi_pdf_uncertainty.py

        returns [numens x space.flat] or [numens]

    """

    ledat = le.load_LEdata(fdict,casename,timesel=timesel, 
                           rettype='ndarray',conv=conv,ftype=ftype,orig=orig,local=local,verb=verb)

    # time needs to be first dimension
    try:
        if ledat.ndim==2:
            ledat = ledat.T
        elif ledat.ndim==3:
            ledat = np.transpose(ledat,(1,0,2))
        else:
            print 'Loaded data is not 2 or 3 dimensions. Do not understand.'
            raise Exception
    except:
        raise

    if seas==None and mo==0:
        lesea=ledat        
    else:
    
        lesea = cutl.seasonalize(ledat,season=seas,mo=mo) # time x numens x space.flat

    if calc!=None and calc=='trend':
        
        mm,bb=cutl.trend_monthly(lesea) # should return 12 x numens x space.flat
        return lesea,mm,bb
    else:
    
        return lesea

```

```python
fdict = {'field': field+region, 'ncfield': ncfield+region, 'comp': comp}
fdictz = {'field': fieldz+regionz, 'ncfield': ncfieldz, 'comp': compz}

# load SCE
lesce1 = load_field(fdict, 'historical', timesel1,mo=mo, verb=False,
                    ftype='fullts', orig=None,local=local)
if seaz in ('DJF','NDJ'):
    lesce1 = lesce1[:-1,:] # get rid of last october (or whatever mo is) to match shorter num of winters.
    
# load "NAM index"
lez1 = load_field(fdictz, 'historical', timesel1,seas=seaz, verb=False,
                    ftype='fullts', orig=None,local=local)

# load SCE
lesce2 = load_field(fdict, 'historical', timesel2,mo=mo, verb=False,
                    ftype='fullts', orig=None,local=local)
if seaz in ('DJF','NDJ'):
    lesce2 = lesce2[:-1,:] # get rid of last october (or whatever mo is) to match shorter num of winters.
    
# load "NAM index"
lez2 = load_field(fdictz, 'historical', timesel2,seas=seaz, verb=False,
                    ftype='fullts', orig=None,local=local)
```

```python
# HISTORICALNAT

# load SCE
lesce1n = load_field(fdict, 'historicalNat', timesel1,mo=mo, verb=False,
                    ftype='fullts', orig=None,local=local)
if seaz in ('DJF','NDJ'):
    lesce1n = lesce1n[:-1,:] # get rid of last october (or whatever mo is) to match shorter num of winters.
    
# load "NAM index"
lez1n = load_field(fdictz, 'historicalNat', timesel1,seas=seaz, verb=False,
                    ftype='fullts', orig=None,local=local)

# load SCE
lesce2n = load_field(fdict, 'historicalNat', timesel2,mo=mo, verb=False,
                    ftype='fullts', orig=None,local=local)
if seaz in ('DJF','NDJ'):
    lesce2n = lesce2n[:-1,:] # get rid of last october (or whatever mo is) to match shorter num of winters.
    
# load "NAM index"
lez2n = load_field(fdictz, 'historicalNat', timesel2,seas=seaz, verb=False,
                    ftype='fullts', orig=None,local=local)
```

```python
# remove climo anomalies
lesce1 = lesce1 - lesce1.mean(axis=0)
lesce2 = lesce2 - lesce2.mean(axis=0)

lez1 = lez1 - lez1.mean(axis=0)
lez2 = lez2 - lez2.mean(axis=0)


lesce1n = lesce1n - lesce1n.mean(axis=0)
lesce2n = lesce2n - lesce2n.mean(axis=0)

lez1n = lez1n - lez1n.mean(axis=0)
lez2n = lez2n - lez2n.mean(axis=0)
```

```python
print lesce1.shape, lez1.shape # time x ensemble
print lesce2.shape, lez2.shape # time x ensemble
```

```python
# HISTORICAL
mm1,bb1,rr1,pp1 = cutl.regress(lesce1,lez1) # mm, bb, r-val, p-val
print mm1.shape
mm2,bb2,rr2,pp2 = cutl.regress(lesce2,lez2) # mm, bb, r-val, p-val

# detrend:
lesce1dtr = cutl.detrend(lesce1,axis=0)
lez1dtr = cutl.detrend(lez1,axis=0)
mm1dtr,bb1dtr,rr1dtr,pp1dtr = cutl.regress(lesce1dtr,lez1dtr) # mm, bb, r-val, p-val

lesce2dtr = cutl.detrend(lesce2,axis=0)
lez2dtr = cutl.detrend(lez2,axis=0)
mm2dtr,bb2dtr,rr2dtr,pp2dtr = cutl.regress(lesce2dtr,lez2dtr) # mm, bb, r-val, p-val
```

```python
# HISTORICALNAT
mm1n,bb1n,rr1n,pp1n = cutl.regress(lesce1n,lez1n) # mm, bb, r-val, p-val
print mm1n.shape
mm2n,bb2n,rr2n,pp2n = cutl.regress(lesce2n,lez2n) # mm, bb, r-val, p-val

# detrend:
lesce1ndtr = cutl.detrend(lesce1n,axis=0)
lez1ndtr = cutl.detrend(lez1n,axis=0)
mm1ndtr,bb1ndtr,rr1ndtr,pp1ndtr = cutl.regress(lesce1ndtr,lez1ndtr) # mm, bb, r-val, p-val

lesce2ndtr = cutl.detrend(lesce2n,axis=0)
lez2ndtr = cutl.detrend(lez2n,axis=0)
mm2ndtr,bb2ndtr,rr2ndtr,pp2ndtr = cutl.regress(lesce2ndtr,lez2dtr) # mm, bb, r-val, p-val
```
