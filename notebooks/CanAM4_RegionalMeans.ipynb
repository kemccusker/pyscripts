```python
import cccmaplots as cplt
import cccmautils as cutl
import cccmaNC as cnc
import constants as con
import cccmacmaps as ccm
import numpy.ma as ma
import pandas as pd
from matplotlib import gridspec

%matplotlib inline

con = reload(con)
printtofile=False

model = 'CanAM4'
bp=con.get_basepath()
basepath=bp['basepath'] + model + '/'; subdir=bp['subdir']
threed=False


field='gz' 
level = 50000 # for threed vars
#latlims=[70,89]; lonlims=[0,359]; region='polcap70' # Polar cap north of 70N
#latlims=[65,89]; lonlims=[0,359]; region='polcap65' # Polar cap north of 65N for NAM proxy
latlims=[60,89]; lonlims=[0,359]; region='polcap60' # Polar cap north of 60N to match pattern corrs
#latlims=[35,60]; lonlims=[40,120]; region='eurasia' # Eurasia 35-60N, 40E-120E
#latlims=[35,60]; lonlims=[240,280]; region='ntham' # North America 35-60N, 120W-80W
#latlims=[35,60]; lonlims=[300,360]; region='nthatl' # North Atlantic 35-60N, 60W-0

casenamec='kemctl1'
casenamep='kem1pert2'

#casenamec='kemnsidcctl'
#casenamep='kemnsidcpert'

casenamec='kemctl1'
casenamep='kem1rcp85a'

if field in ('gz','u','t','v','q'):
    threed=True
    fldstr = field + str(level/100) # for figure names
    field = field + str(level) # for filename
else:
    fldstr=field
    ncfield=field.upper()
    


if fldstr=='st':
    ylims = ((-3,3), (-4,3), (-3,3), (-2,2))
    if region in ('polcap60','polcap65','polcap70'):
        clims = -3,3
    else:
        clims = -1,1
            
    conv=1
elif fldstr=='pmsl':
    ylims = ((-6,6),(-6,6),(-4,5),(-4,4))
    clims=-1,1
    conv=1
elif fldstr=='gz500':
    ylims = ((-70,70),(-100,100),(-80,80),(-50,50))
    clims=-25,25
    ncfield='PHI'
    conv = 1/con.get_g()
        

timesel = '0002-01-01,0121-12-31'
seasons = ('SON','DJF','MAM','JJA')

fnamec = basepath + casenamec + subdir + casenamec + '_' + field + '_001-121_ts.nc'
fnamep = basepath + casenamep + subdir + casenamep + '_' + field + '_001-121_ts.nc'
time = cnc.getNCvar(fnamec,'time',timesel=timesel)
ntime = len(time)

lat = cnc.getNCvar(fnamec,'lat')
lon = cnc.getNCvar(fnamec,'lon')
lons,lats = np.meshgrid(lon,lat)

reglatsbool = np.logical_and(lat>latlims[0],lat<latlims[1])
reglonsbool = np.logical_and(lon>lonlims[0],lon<lonlims[1])
"""reglats = lat[reglatsbool]
reglons = lon[reglonsbool]"""

# testing
fldcsea1 = cnc.getNCvar(fnamec,ncfield,timesel=timesel,seas='DJF')*conv
# try masking where we don't want (outside Eurasia in this case)
regmask = np.logical_or( 
                        np.logical_or(lats<latlims[0],lats>latlims[1]), 
                        np.logical_or(lons<lonlims[0],lons>lonlims[1]))
regmaskt = np.tile(regmask,(119,1,1))
fldcsea1m = ma.masked_where(regmaskt,fldcsea1)

areas = cutl.calc_cellareas(lat,lon)
areasm = ma.masked_where(regmask,areas)
weightsm = areasm / np.sum(np.sum(areasm,axis=1),axis=0) # I think this will only sum non-masked values

# what does the region look like?
plt.figure()
cplt.kemmap(areasm,lat,lon,type='nh')
```

```python
# seasonal
fldcseadict = {}
fldpseadict = {}
flddseadict = {}
flddsearegdict = {}
fldcsearegdict = {}
fldpsearegdict = {}
flddseamdict = {}
fldcseamdict = {}
fldpseamdict = {}

for seaii,sea in enumerate(seasons):
    
    fldcseadict[sea] = cnc.getNCvar(fnamec,ncfield,timesel=timesel,seas=sea)*conv
    fldpseadict[sea] = cnc.getNCvar(fnamep,ncfield,timesel=timesel,seas=sea)*conv
    
    flddseadict[sea] = fldpseadict[sea]-fldcseadict[sea]
    # regional subset
    flddsea = flddseadict[sea]
    ntime = flddsea.shape[0]
    tmp = flddsea[:,reglatsbool,:] # this swaps the lat and time dims for some reason
    tmp = np.transpose(tmp,(1,0,2))
    tmp = tmp[:,:,reglonsbool]
    tmp = np.transpose(tmp,(1,0,2))
    
    #flddsearegdict[sea] = tmp
    
    # regional mask
    regmaskt = np.tile(regmask,(ntime,1,1))
    flddseamdict[sea] = ma.masked_where(regmaskt,flddsea)
    fldcseamdict[sea] = ma.masked_where(regmaskt,fldcseadict[sea])
    fldpseamdict[sea] = ma.masked_where(regmaskt,fldpseadict[sea])
    

    weightsmt = np.tile(weightsm,(ntime,1,1))
    flddsearegdict[sea] = np.sum(np.sum(flddseamdict[sea]*weightsmt,axis=2),axis=1)
    fldcsearegdict[sea] = np.sum(np.sum(fldcseamdict[sea]*weightsmt,axis=2),axis=1)
    fldpsearegdict[sea] = np.sum(np.sum(fldpseamdict[sea]*weightsmt,axis=2),axis=1)

```

```python
# Plot just the timeseries with running means. Test axis limits (no ylims set)
N=10
clr='k'
fig, axs = plt.subplots(1,4)
fig.set_size_inches((16,4))

for sii,sea in enumerate(seasons):

    ntime=len(fldcsearegdict[sea])
    ax=axs[sii]

    ax.plot(fldcsearegdict[sea]-np.mean(fldcsearegdict[sea]),clr)
    ax.plot(fldpsearegdict[sea]-np.mean(fldcsearegdict[sea]),clr+'--')

    xxrun=np.arange((N-1)/2.,ntime-(N-1)/2.)

    runc = np.convolve(fldcsearegdict[sea]-np.mean(fldcsearegdict[sea]), np.ones((N,))/N, mode='valid')
    ax.plot(xxrun,runc, 'k',linewidth=2)
    runp = np.convolve(fldpsearegdict[sea]-np.mean(fldcsearegdict[sea]), np.ones((N,))/N, mode='valid')
    ax.plot(xxrun,runp, 'k',linewidth=3)
    ax.fill_between(xxrun, runc, runp, where=runp>runc, facecolor='red', alpha=0.5)
    ax.fill_between(xxrun, runc, runp, where=runp<runc, facecolor='blue', alpha=0.5)
    ax.set_title(sea)
    #ax.set_ylim(ylims[sii])

reg_ax = fig.add_axes([.04,.13, .23,.23])
cplt.kemmap(areasm*0,lat,lon,axis=reg_ax,type='nh',latlim=30,suppcb=1,cmap='blue2blue_w10')

# Don't save. Use as a ylims tester
#if printtofile:
#    fig.savefig(fldstr + 'diff_' + casenamep + '_v_' + casenamec + '_allseas_' +\
#                region + '_timeseries_run' + str(N) + '.pdf')
```
