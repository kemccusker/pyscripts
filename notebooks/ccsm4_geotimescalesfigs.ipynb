```python
# convert estimated melt rate (m/yr) to mm/yr SLR. 
# SLR = addedvolume / oceansfcarea
# addedvolume = (m/yr * iceshelfarea) --> m3/yr --> metric tonnes/yr --> 1000kg/yr
#     1Gt = 10^9 tonnes = 10^9 m3 water = 1km3 water
#     
# oceansfcarea = 361,900,000 km2
#    from http://www.ngdc.noaa.gov/mgg/global/etopo1_ocean_volumes.html
#
# Need area of ice shelf @@
# Need estimated basal melt rate from temp anomaly
# Rule of thumb conversion factor from Rignot et al 2002: 
#    0.1C = 1m/yr melt rate
#
# example reprocessing of data to get just PIG region and SH:
# cdo selindexbox,230,280,1,187 TEMP.geo2035ensavg.pop.h.2094.nc TEMP.geo2035ensavg.pop.h.2094.PIG.nc
#
# Ice Shelf Area (ISA) taken from Rignot 2002 (data from 1996): 
# PIG: 795, THW: 413, SMI: 341, KOH: 210, DVQ: 384, LAN: 167  [km2]
# "the melting points for the peak ice thickness at the grounding 
#    line (Hp) drafts in this study range from –2.6 to –4.0°C."
#  @@ Need to take a difference from the melt point?
# Rignot 2002 quotes grounding lines of PIG ~1km, which is deeper than I have seen elsewhere.
#  avg groundine line depth: PIG: 1.0, THW: 0.9, SMI: 1.0, KOH: 1.1, DVQ: 0.8, LAN: 0.9
#    Is it the differenc b/w ice shelf base and grounding line? Yes, see Jenkins et al
# @@@ Joughin et al 2014 has Thwaites grounding line at 600m (from Holt et al 2006)
# Vaughan et al 2006 has PIG at ~1000m 
#    (and actually Holt et al *looks* like Thwaites is about 8-900m)
# @@@@ Now I just need melting point of seawater at that pressure and salinity...
#    or, assume it is already above freeze point, then just need anomaly from control. @@

import cccmautils as cutl
import numpy.ma as ma
import pandas as pd

%matplotlib inline

pigisa={'PIG': 795, 'THW': 413, 'SMI': 341, 'KOH': 210, 'DVQ': 384, 'LAN': 167}#km2
piggld={'PIG': 1.0, 'THW': 0.9, 'SMI': 1.0, 'KOH': 1.1, 'DVQ': 0.8, 'LAN': 0.9}#km
pigisasr=pd.Series(pigisa)
piggldsr=pd.Series(piggld)

totisa=pigisasr.sum()
print totisa
km2tom2=1000*1000
totisa=totisa*km2tom2

isawgts=pigisasr/np.float(totisa)
print isawgts

avggld = np.average(piggldsr,weights=isawgts)
print avggld
#avggldnowgt = piggldsr.mean() # basically the same
#print avggldnowgt

# because we are doing large-scale analysis and the base of ice shelves starts around 400m,
#   let's average from 400-800m?
```

```python
import numpy.ma as ma

%matplotlib inline

#basepath='/Volumes/MyPassport2TB/DATA/ccsm4/'
casename1='geo2035ensavg'
casename2='rcp8_5GHGrem1850'
#subdir='/ocn_proc/'

basepath = '/Users/kelly/School/DATA/'

casenamec = 'b40.20th.track1.1deg.006'
field = 'TEMP'
years='2035-2094'
years2='2035-2081'

fnamec = basepath + casenamec + '/' + casenamec + '.pop.ANN.1970-1999.nc'
fname1 = basepath + casename1 + '/' + field + '.' + casename1 + '.pop.'+years+'.PIG.nc'
fname2 = basepath + casename2 + '/' + field + '.' + casename2 + '.pop.'+years2+'.PIG.nc'

# for regular TLAT/TLON
#fname=basepath+casename1+subdir+field+'/'+field+'.'+casename1+'.pop.h.2094.nc' 
tlat=cnc.getNCvar(fnamec,'TLAT') # full coords
tlon=cnc.getNCvar(fnamec,'TLONG') # full coords
lev=cnc.getNCvar(fnamec,'z_t')
kmt = cnc.getNCvar(fnamec,'KMT') # k Index of deepest grid cell on T grid
tarea = cnc.getNCvar(fnamec,'TAREA')

piglon = range(230,281) # lon indices
piglat = range(0,187) # lat indices

#fname=basepath+casename1+subdir+field+'/'+field+'.'+casename1+'.pop.'+years+'.PIG.nc'
print fname1
tlatp = cnc.getNCvar(fname1,'TLAT')
tlonp=cnc.getNCvar(fname1,'TLONG')
zt=cnc.getNCvar(fname1,'z_t')
fld1=cnc.getNCvar(fname1,'TEMP') # PIG region only
print fld1.shape
print fname2
fld2=cnc.getNCvar(fname2,'TEMP') # PIG region only
print fld2.shape

#fname='/Volumes/MyPassport2TB/DATA/ccsm4/geo2035ensavg/ocnclimo/geo2035ensavg.pop.ANN.2045-2054.nc'
#htn = cnc.getNCvar(fname,'HTN') # cell widths on north sides of T cells
#hte = cnc.getNCvar(fname,'HTE') # cell widths on east sides of T cells
#kmt = cnc.getNCvar(fname,'KMT') # k Index of deepest grid cell on T grid
#tarea = cnc.getNCvar(fname,'TAREA')

kmtpig=kmt[piglat,:]
kmtpig=kmtpig[:,piglon]
tareapig=tarea[piglat,:]
tareapig=tareapig[:,piglon]

print kmtpig.shape
# tile kmt with time for fld1
kmtpigt1 = np.tile(kmtpig,(fld1.shape[0],1,1))
print kmtpigt1.shape
# tile kmt with time for fld2
kmtpigt2 = np.tile(kmtpig,(fld2.shape[0],1,1))
print kmtpigt2.shape

tareapigt = np.tile(tareapig,(len(zt),1,1))
tareapigtt1 = np.tile(tareapigt,(fld1.shape[0],1,1,1)) # add time dim for fld1
tareapigtt2 = np.tile(tareapigt,(fld2.shape[0],1,1,1)) # add time dim for fld2

tareat = np.tile(tarea,(len(zt),1,1))
print tareat.shape

temparea=ma.zeros(tareat.shape)
tempfldpig1=ma.zeros(fld1.shape)
tempareapig1=ma.zeros(fld1.shape)
tempfldpig2=ma.zeros(fld2.shape)
tempareapig2=ma.zeros(fld2.shape)

for lii,zz in enumerate(zt):
    # mask out levels below sea floor
    temparea[lii,...] = ma.masked_where(kmt <= lii,tareat[lii,...]) # this is full area
    
    tempfldpig1[:,lii,...] = ma.masked_where(kmtpigt1 <= lii,fld1[:,lii,...]) # a time dimension here
    tempareapig1[:,lii,...] = ma.masked_where(kmtpigt1 <= lii, tareapigtt1[:,lii,...])# time dim too
    tempfldpig2[:,lii,...] = ma.masked_where(kmtpigt2 <= lii,fld2[:,lii,...]) # a time dimension here
    tempareapig2[:,lii,...] = ma.masked_where(kmtpigt2 <= lii, tareapigtt2[:,lii,...])# time dim too

# what is correct grid cell area? is there an ocean fraction? 
# I don't think so. just have to check for depth of deepest cell
# when doing region mean, need to use tarea and total tarea in lat circle...
#  (don't need for zonal mean tho)

# zonal mean:
tempfldpigzm1=ma.mean(tempfldpig1,axis=3)
print 'tempfldpigzm1.shape: ' + str(tempfldpigzm1.shape)
tempfldpigzm2=ma.mean(tempfldpig2,axis=3)
print 'tempfldpigzm2.shape: ' + str(tempfldpigzm2.shape)
```

```python
Nlim=-72   # 65 to 74 is good for pig
Slim=-74

# Now pick the sublat range from tareapig and create weights
tareapigsub1 = tempareapig1[:,:,np.logical_and(onelon<= Nlim,onelon>Slim),:] # time x depth x sublat x lon
tareapigsub2 = tempareapig2[:,:,np.logical_and(onelon<= Nlim,onelon>Slim),:] # time x depth x sublat x lon
##  total area in sublat x lon box
totareapigxy1 = ma.sum(ma.sum(tareapigy1,axis=2),axis=2) # time x depth 
totareapigxy2 = ma.sum(ma.sum(tareapigy2,axis=2),axis=2) # time x depth

totareapigx1 = ma.sum(tareapigsub1,axis=3) # total area in the lon range for e/ time x depth x lat
totareapigx2 = ma.sum(tareapigsub2,axis=3) # total area in the lon range for e/ time x depth x lat

# should be able to take zonal mean no prob
#   then average in meridional direction with weights totareapigx / totareapigxy per latitude
# zm already done: tempfldpigzm1,2

# first tile over sublat to create weights
totareapigxyt2 = np.tile(totareapigxy2,(tareapigsub2.shape[2],1,1)) 
totareapigxyt2=np.transpose(totareapigxyt2,(1,2,0)) # time x depth x lat?
wgtspig2 = totareapigx2 / totareapigxyt2

totareapigxyt1 = np.tile(totareapigxy1,(tareapigsub1.shape[2],1,1)) 
totareapigxyt1=np.transpose(totareapigxyt1,(1,2,0)) # time x depth x lat?
wgtspig1 = totareapigx1 / totareapigxyt1
print 'wgtspig1.shape: ' + str(wgtspig1.shape)
print 'wgtspig2.shape: ' + str(wgtspig2.shape)
print 'tempfldpigzm1.shape: ' + str(tempfldpigzm.shape)

# then average meridionally
avgpig1,ret1 = ma.average(tempfldpigzm1[:,:,np.logical_and(onelon<= Nlim,onelon>Slim)],
                     axis=2,weights=wgtspig1,returned=True)
avgpig2,ret2 = ma.average(tempfldpigzm2[:,:,np.logical_and(onelon<= Nlim,onelon>Slim)],
                     axis=2,weights=wgtspig2,returned=True)

print avgpig1.shape
```

```python
# convert estimated melt rate (m/yr) to mm/yr SLR. 
# SLR = addedvolume / oceansfcarea
# addedvolume = (m/yr * iceshelfarea) --> m3/yr --> metric tonnes/yr --> 1000kg/yr
#     1Gt = 10^9 tonnes = 10^9 m3 water = 1km3 water
#     
# oceansfcarea = 361,900,000 km2

ocnsa = 361900000 # km2

# Rule of thumb conversion factor from Rignot et al 2002: 
#    0.1C = 1m/yr melt rate
basalm1=avg1*10 # C * m/yr
vol1 = basalm1*totisa
basalm2=avg2*10 # C * m/yr
vol2 = basalm2*totisa

plt.figure()
plt.plot(vol1/1e9,'b')
plt.plot(vol2/1e9,'g')
plt.title('Gt/yr or km3/yr')

SLR1=vol1/1e9 / ocnsa
SLR2=vol2/1e9 / ocnsa

kmtomm=1000000
plt.figure()
plt.plot(SLR1*kmtomm,'b')
plt.plot(SLR2*kmtomm,'g')
plt.title('SLR (mm)')

```
