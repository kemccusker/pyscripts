```python
# convert estimated melt rate (m/yr) to mm/yr SLR. 
# SLR = addedvolume / oceansfcarea
# addedvolume = (m/yr * iceshelfarea) --> m3/yr --> metric tonnes/yr --> 1000kg/yr
#     1Gt = 10^9 tonnes = 10^9 m3 water = 1km3 water
#     
# oceansfcarea = 361,900,000 km2
#    from http://www.ngdc.noaa.gov/mgg/global/etopo1_ocean_volumes.html
#
# Need area of ice shelf @@
# Need estimated basal melt rate from temp anomaly
# Rule of thumb conversion factor from Rignot et al 2002: 
#    0.1C = 1m/yr melt rate
#
# example reprocessing of data to get just PIG region and SH:
# cdo selindexbox,230,280,1,187 TEMP.geo2035ensavg.pop.h.2094.nc TEMP.geo2035ensavg.pop.h.2094.PIG.nc
#
# Ice Shelf Area (ISA) taken from Rignot 2002 (data from 1996): 
# PIG: 795, THW: 413, SMI: 341, KOH: 210, DVQ: 384, LAN: 167  [km2]
# "the melting points for the peak ice thickness at the grounding 
#    line (Hp) drafts in this study range from –2.6 to –4.0°C."
#  @@ Need to take a difference from the melt point?
# Rignot 2002 quotes grounding lines of PIG ~1km, which is deeper than I have seen elsewhere.
#  avg groundine line depth: PIG: 1.0, THW: 0.9, SMI: 1.0, KOH: 1.1, DVQ: 0.8, LAN: 0.9
#    Is it the differenc b/w ice shelf base and grounding line? Yes, see Jenkins et al
# @@@ Joughin et al 2014 has Thwaites grounding line at 600m (from Holt et al 2006)
# Vaughan et al 2006 has PIG at ~1000m 
#    (and actually Holt et al *looks* like Thwaites is about 8-900m)
# @@@@ Now I just need melting point of seawater at that pressure and salinity...
#    or, assume it is already above freeze point, then just need anomaly from control. @@

import cccmautils as cutl
import numpy.ma as ma
import pandas as pd

%matplotlib inline

pigisa={'PIG': 795, 'THW': 413, 'SMI': 341, 'KOH': 210, 'DVQ': 384, 'LAN': 167}#km2
piggld={'PIG': 1.0, 'THW': 0.9, 'SMI': 1.0, 'KOH': 1.1, 'DVQ': 0.8, 'LAN': 0.9}#km
pigisasr=pd.Series(pigisa)
piggldsr=pd.Series(piggld)

totisa=pigisasr.sum()
print totisa

isawgts=pigisasr/np.float(totisa)
print isawgts

avggld = np.average(piggldsr,weights=isawgts)
print avggld
#avggldnowgt = piggldsr.mean() # basically the same
#print avggldnowgt

# because we are doing large-scale analysis and the base of ice shelves starts around 400m,
#   let's average from 400-800m?
```
