```python
import cccmautils as cutl

# Finally, average in the vertical
dz=np.diff(zt) # cm, same as TAREA

#print zt/100. # m # depth of midpoint of layer

UB=20000; UBidx=cutl.find_nearest(zt,UB) # 300m upper bound
LB=80000; LBidx=cutl.find_nearest(zt,LB) # 800m (787) lower bound
print 'upper bound: ' + str(zt[UBidx]/100.) + 'm'
print 'lower bound: ' + str(zt[LBidx]/100.) + 'm'

dzsub=dz[UBidx:LBidx+1]
totdz = ma.sum(dzsub)
dzwgts=dzsub / np.float(totdz)
# these weights should be good for control too
dzwgtst1=np.tile(dzwgts,(avgpig1.shape[0],1))
dzwgtst2=np.tile(dzwgts,(avgpig2.shape[0],1))

# CONTROL: interp layers to midpoints and get sublayers
avgpigct1shift = (avgpigct1[:,0:-1]+avgpigct1[:,1:]) / 2.
avgpigct1shift=avgpigct1shift[:,UBidx:LBidx+1]
avgc1=ma.average(avgpigct1shift,axis=1,weights=dzwgtst1)

avgc=avgc1[0] # just need a scalar now


# PERTS: interp layers to midpoints and get sublayers
avgpig1shift = (avgpig1[:,0:-1]+avgpig1[:,1:]) / 2.
avgpig2shift = (avgpig2[:,0:-1]+avgpig2[:,1:]) / 2.
avgpig1shift=avgpig1shift[:,UBidx:LBidx+1]
avgpig2shift=avgpig2shift[:,UBidx:LBidx+1]


print dzwgtst2.shape
print avgpig1shift.shape

avg1=ma.average(avgpig1shift,axis=1,weights=dzwgtst1)
avg2=ma.average(avgpig2shift,axis=1,weights=dzwgtst2)

plt.figure()
plt.plot(avg1,'b')
plt.plot(avg2,'g')
plt.plot(avgc1,'k')
plt.title('avg PIG T')
print 'avgc: ' + str(avgc)
```

```python
# convert estimated melt rate (m/yr) to mm/yr SLR. 
# SLR = addedvolume / oceansfcarea
# addedvolume = (m/yr * iceshelfarea) --> m3/yr --> metric tonnes/yr --> 1000kg/yr
#     1Gt = 10^9 tonnes = 10^9 m3 water = 1km3 water
#     
# oceansfcarea = 361,900,000 km2

ocnsa = 361900000 # km2

# Rule of thumb conversion factor from Rignot et al 2002: 
#    0.1C = 1m/yr melt rate

# CONTROL
basalmc=avgc*10
volc=basalmc*totisa
print  'volc: ' + str(volc)

basalm1=avg1*10 # C * m/yr
vol1 = basalm1*totisa # m3/yr (or tonne/yr or 1000kg/yr)
basalm2=avg2*10 # C * m/yr
vol2 = basalm2*totisa


plt.figure()
plt.plot(vol1/1e9,'b')
plt.plot(vol2/1e9,'g')
plt.axhline(y=volc/1e9,color='k')
plt.title('Gt/yr or km3/yr')

SLR1=vol1/1e9 / ocnsa # km
SLR2=vol2/1e9 / ocnsa # km
SLRc=volc/1e9 / ocnsa
print 'SRLc: ' + str(SLRc)

kmtomm=1000000

plt.figure()
plt.plot(SLR1*kmtomm,'b')
plt.plot(SLR2*kmtomm,'g')
plt.axhline(y=SLRc*kmtomm,color='k')
plt.title('SLR (mm)')
plt.xlabel(str(Nlim) + '-' + str(Slim) +', ' + str(UB/100) + 'm-' + str(LB/100))


```

```python
# Calc as anomaly from melt point:
mpstd=-1.8 # melt point of seawater (not considering pressure)
mpa=-2.6 # range from Rignot 2002
mpb=-4   # range from Rignot 2002

# std melt point
basalmstdc=(avgc-mpstd)*10
volstdc = basalmstdc*totisa

basalmstd1=(avg1-mpstd)*10 # C * m/yr
volstd1 = basalmstd1*totisa # m3/yr (or tonne/yr or 1000kg/yr)
basalmstd2=(avg2-mpstd)*10 # C * m/yr
volstd2 = basalmstd2*totisa
# warm end of range
basalmac=(avgc-mpa)*10 # C * m/yr
volac = basalmac*totisa

basalma1=(avg1-mpa)*10 # C * m/yr
vola1 = basalma1*totisa
basalma2=(avg2-mpa)*10 # C * m/yr
vola2 = basalma2*totisa
# cold end of range
basalmbc=(avgc-mpb)*10 # C * m/yr
volbc = basalmbc*totisa

basalmb1=(avg1-mpb)*10 # C * m/yr
volb1 = basalmb1*totisa
basalmb2=(avg2-mpb)*10 # C * m/yr
volb2 = basalmb2*totisa

SLRstd1=volstd1/1e9 / ocnsa # km
SLRstd2=volstd2/1e9 / ocnsa # km

SLRa1=vola1/1e9 / ocnsa # km
SLRa2=vola2/1e9 / ocnsa # km
SLRb1=volb1/1e9 / ocnsa # km
SLRb2=volb2/1e9 / ocnsa # km

# control SLR
SLRstdc=volstdc/1e9/ocnsa
SLRac=volac/1e9/ocnsa
SLRbc=volbc/1e9/ocnsa



plt.figure()
plt.axhline(volstdc/1e9,color='k',linewidth=2)
plt.plot(volstd1/1e9,'b',linewidth=2)
plt.plot(volstd2/1e9,'g',linewidth=2)
plt.axhline(volac/1e9,color='k',linewidth=1)
plt.plot(vola1/1e9,'b',linewidth=1)
plt.plot(vola2/1e9,'g',linewidth=1)
plt.axhline(volbc/1e9,color='k',linewidth=1,linestyle='--')
plt.plot(volb1/1e9,'b',linewidth=1,linestyle='--')
plt.plot(volb2/1e9,'g',linewidth=1,linestyle='--')
plt.title('Gt/yr (km3/yr): meltpt: -1.8C (thick), -2.6C (thin), -4C (dash)')
plt.xlabel(str(Nlim) + '-' + str(Slim) +', ' + str(UB/100) + 'm-' + str(LB/100)+'m')



kmtomm=1000000
plt.figure()
plt.axhline(SLRstdc*kmtomm,color='k',linewidth=2)
plt.plot(SLRstd1*kmtomm,'b',linewidth=2)
plt.plot(SLRstd2*kmtomm,'g',linewidth=2)
plt.axhline(SLRac*kmtomm,color='k',linewidth=1)
plt.plot(SLRa1*kmtomm,'b',linewidth=1)
plt.plot(SLRa2*kmtomm,'g',linewidth=1)
plt.axhline(SLRbc*kmtomm,color='k',linewidth=1,linestyle='--')
plt.plot(SLRb1*kmtomm,'b',linewidth=1,linestyle='--')
plt.plot(SLRb2*kmtomm,'g',linewidth=1,linestyle='--')
plt.title('SLR (mm) meltpt: -1.8C (thick), -2.6C (thin), -4C (dash)')
plt.xlabel(str(Nlim) + '-' + str(Slim) +', ' + str(UB/100) + 'm-' + str(LB/100)+'m')
```
