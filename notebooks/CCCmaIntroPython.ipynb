```python
# Goals: Read data from a netcdf file, do some averaging, make some figures

import numpy as np                # import numpy to work with arrays/matrices
import matplotlib.pyplot as plt   # import matplotlib to do plotting
# netCDF4 package to manipulate netCDF files.
# Not included in basic install but Mike B. installed it for everyone.
from netCDF4 import Dataset       # https://code.google.com/p/netcdf4-python/      
from mpl_toolkits.basemap import Basemap # Basemap for plotting maps

# a 'magic' function to show figures inline in the notebook
%matplotlib inline


# Step one: read in some data from an AGCM simulation named 'kemnsidccctl'


# set up the filepath
#basepath = '/Volumes/MyPassport2TB/DATA/CanSISE/CanAM4/'; subdir='' # personal laptop
basepath = '/home/rkm/work/DATA/CanAM4/'; subdir='ts/'            # work linux
runname='kemnsidcctl'
field = 'st'; ncfield='ST' # variable name
timeper = '001-121'

fname = runname + '_' + field + '_' + timeper + '_ts.nc' # '+' concatenates strings
ffull = basepath + runname + '/' + subdir + fname

print 'filename: ' + ffull # print full file path
```

```python
# open the file using netCDF4 Dataset
ncfile = Dataset(ffull,'r')
print type(ncfile)
#help(ncfile)

#print ncfile.CCCma_runid
runid = ncfile.getncattr('CCCma_runid')
print runid
""" this is my documentation comment, docstring
     it is multi-line
     this comes out when calling help() on objects
"""
```

```python
# len() is a python built-in length function. 
#'dimensions' is an attribute of ncfile, which is an object of type Dataset
# the result is number of dimensions in the ncfile, which we set to ndims.
dims = ncfile.dimensions
print dims
print type(dims)

ndims = len(ncfile.dimensions) 
print ndims
```

```python
# get coordinate data
# get the 'lat' variable as a numpy.ndarray 
# [:] makes it an array. otherwise it's a netCDF4.Variable object
latobj = ncfile.variables['lat']
print latobj

lat = ncfile.variables['lat'][:] 
lon = ncfile.variables['lon'][:]

print lat
print type(lat)
# type() is a built-in function that gives the datatype
# str() is a built-in function that turns an object into a string
#    otherwise get 'TypeError: cannot concatenate 'str' and 'type' objects'
print 'Datatype of lat variable: ' + str(type(lat)) 
```

```python
# lat is a numpy.ndarray which has attribute 'shape' (like matlab size())
# use str() to turn shape into a string
# Otherwise get 'TypeError: cannot concatenate 'str' and 'tuple' objects'
print 'Shape of lat var: ' + str(lat.shape)
```

```python
# get the data (here, screen temperature)
fldobj = ncfile.variables[ncfield]
fld = ncfile.variables[ncfield][...] # the ellipsis means get all of the data no matter how many dims
print fld.shape # can print a number if not combining with a string
print type(fldobj)
print fldobj.dimensions
```

```python
# fld is now an array of temperature w/ dimensions: time (months) x lat x lon

fldtm = np.mean(fld,axis=0) # take time-mean of first axis. INDEXING IS ZERO-BASED
shlat = lat[lat<0]
print shlat
fldsub = np.mean(fld[:,lat<0,:],axis=0) # example of indexing
print fldsub.shape
print fldtm.shape
print fldtm[0:11,:].shape # index range is NOT inclusive of last index
```

```python
# Plotting, the best part of science!
#  matplotlib is very similar to matlab plotting

plt.figure()
plt.plot(fld[:,50,100]) # choose one random grid cell to plot in time
plt.xlabel('time')
plt.ylabel(field)
```

```python
# Now for some maps
# Basemap is a handy package that makes plotting maps easy
#    http://matplotlib.org/basemap/
bm = Basemap(projection='robin',lon_0=180,lat_0=0, resolution='c')

# pcolor(), contour(), contourf() need a mesh of coords
lons, lats = np.meshgrid(lon,lat)

# filled contours
# 20 auto levels. latlon=True means the coords given are lat/lon, not x/y
pc = bm.contourf(lons,lats,fldtm,20,latlon=True) 
bm.drawcoastlines(color='0.7') # add coastlines

```

```python
# make the same map with a different projection and add more features
#   see many examples here: http://matplotlib.org/basemap/users/examples.html

bm = Basemap(projection='ortho',lat_0=45,lon_0=-100,resolution='c')

# filled contours
# 20 auto levels. latlon=True means the coords given are lat/lon, not x/y. 
# vmin/vmax are color axis limits
pc = bm.contourf(lons,lats,fldtm,20,latlon=True,vmin=-35,vmax=35) 
bm.drawcoastlines(color='0.7') # add coastlines

# add lat lon lines
bm.drawmeridians(np.arange(0,360,30)) # np.arange(st,end,step) creates an array of numbers 
bm.drawparallels(np.arange(-90,90,30))

# add a colorbar
cbar = bm.colorbar(pc,location='bottom',pad="5%")
cbar.set_label('C')
```

```python
# more advanced topics

# Above, when creating the Basemap, we set a bunch of key,value pairs (key='value'). 
# We can use a "dictionary" to accomplish this as well.
# Dictionaries are useful for lots of things.....

mapdict = dict(projection='ortho',lat_0=45,lon_0=-100,resolution='c')
# OR another notation
mapdict2 = {'projection': 'ortho',
            'lat_0': 45,
            'lon_0': -100,
            'resolution': 'c' }

print mapdict

print mapdict2
```

```python
# now when creating the Basemap, we can use 'expansion' of the dict
# the ** expands the dictionary into key/value pairs. 
# This can be done for any function that takes key/value inputs, which most do.
bm = Basemap(**mapdict) 

# The same can be done for the params in the contourf() function
# Here I add a parameter for colormap. 
# 'Spectral_r' is a built-in colormap (the reverse of 'Spectral')
# See here for built-in cmaps and how to show them and add more:
# http://wiki.scipy.org/Cookbook/Matplotlib/Show_colormaps
contdict = dict(latlon=True,vmin=-35,vmax=35,cmap='Spectral_r',extend='both')

pc = bm.contourf(lons,lats,fldtm,20,**contdict) # add the key/value pairs after the required args
bm.drawcoastlines(color='0.3') # add coastlines, darker this time

# add lat lon lines
bm.drawmeridians(np.arange(0,360,30)) # np.arange(st,end,step) creates an array of numbers 
bm.drawparallels(np.arange(-90,90,30))

# add a colorbar
cbar = bm.colorbar(pc,location='bottom',pad="5%")
cbar.set_label('C')
```

```python
# Given the above example, you might wonder why the dictionary expansion
# is useful since we just defined the dictionary and then expanded it, 
# putting two steps where there could be one.
# An example is if you don't know at first what type of plot you are 
# building

# A very simple and contrived example:

#   initialize dict to build on
mapdict = {'resolution':'c'} # I want all my maps to have coarse coastlines
contdict = {'latlon': True, 'extend':'both'}


# set my choices for type of figure, so I only have
#  to adjust these settings to create different figures

npolar=True # do I want an Arctic-centered plot or not?
latlim=45 # do I want a southern lat limit? if not, set to None.
cmap='Spectral_r'
cmin=-35; cmax=35

#  Imagine you could put this functionality into a function....

if npolar:
    # Projections in Basemap each require their own set of params
    if latlim!=None:
        mapdict['projection'] = 'npstere' # add this key/val pair
        mapdict['boundinglat'] = latlim   # and this one
        mapdict['lon_0'] = 0.  # the trailing '.' make this integer a float
    else:
        mapdict['projection'] = 'ortho'
        mapdict['lon_0'] = 0.  # the trailing '.' make this integer a float
        mapdict['lat_0'] = 90. # the trailing '.' make this integer a float

else:
    mapdict['projection'] = 'robin'
    mapdict['lon_0']=180.
    mapdict['lat_0'] = 0.  

if cmap!=None:
    contdict['cmap'] = cmap
if cmin !=None and cmax != None:
    contdict['cmin'] = cmin
    contdict['cmax'] = cmax
    

bm = Basemap(**mapdict)
pc = bm.contourf(lons,lats,fldtm,20,**contdict)  
bm.drawcoastlines(color='.3')
cbar = bm.colorbar(pc,location='bottom',pad="5%")
cbar.set_label('C')
plt.title('SAT climo',fontsize=10)
```
