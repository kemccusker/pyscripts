```python
def load_field(fdict,casename,timesel,seas=None,mo=0,ftype='fullts',orig=None,calc=None,conv=1,local=False,verb=False):
    
    """ TAKEN FROM load_field() in canesm_le_composite.py/pi_pdf_uncertainty.py

        returns [numens x space.flat] or [numens]

    """

    ledat = le.load_LEdata(fdict,casename,timesel=timesel, 
                           rettype='ndarray',conv=conv,ftype=ftype,orig=orig,local=local,verb=verb)

    # time needs to be first dimension
    try:
        if ledat.ndim==2:
            ledat = ledat.T
        elif ledat.ndim==3:
            ledat = np.transpose(ledat,(1,0,2))
        else:
            print 'Loaded data is not 2 or 3 dimensions. Do not understand.'
            raise Exception
    except:
        raise

    if seas==None and mo==0:
        lesea=ledat        
    else:
    
        lesea = cutl.seasonalize_monthlyts(ledat,season=seas,mo=mo) # time x numens x space.flat

    if calc!=None and calc=='trend':
        
        mm,bb=cutl.trend_monthly(lesea) # should return 12 x numens x space.flat
        return lesea,mm,bb
    else:
    
        return lesea


```

```python
# from samplecode/monthly_sie_trends_kelly.py

import matplotlib.pyplot as plt
import matplotlib as mpl
plt.ion()
plt.close('all')
import numpy as np
import cmipdata as cd

def get_model_trends(data):
    """ Get the trends from ens for each

       Return trends in /decade
    """
    s = data.shape
    trends = np.zeros((s[0], 12))
    for i in np.arange(s[0]):
        for month in range(12):
            md = data[i, month::12]
            trends[i, month], _ = np.polyfit(np.arange(md.shape[0]), md, 1)
    return trends*10


# Load the Arctic extent data
path_sic = '/raid/ra40/data/ncs/cmip5/sic/'

# make an ensemble object
ens_sic = cd.mkensemble(path_sic + 'area-integral*CanESM2*'
                    , prefix=path_sic + 'area-integral_tmp_')
mat_sic = cd.loadfiles(ens_sic, 'sic',
                       start_date='1979-01-01', end_date='2012-12-31')/1e12
cn = []
for model, experiment, realization, variable, files in ens_sic.iterate():
    cn.append(model.name.replace('-','_') + '_' + realization.name)


# Load the Antarctic extent data
path_sic_ant = '/raid/ra40/data/ncs/cmip5/sic/antarctic/'

# make an ensemble object
ens_sic_ant = cd.mkensemble(path_sic_ant + 'area-integral*CanESM2*'
                    , prefix=path_sic_ant + 'area-integral_tmp_')
mat_sic_ant = cd.loadfiles(ens_sic_ant, 'sic',
                       start_date='1979-01-01', end_date='2012-12-31')/1e12
cn_ant = []
for model, experiment, realization, variable, files in ens_sic_ant.iterate():
    cn_ant.append(model.name.replace('-','_') + '_' + realization.name)

# compute the monthly trends
trends = get_model_trends(mat_sic)
trends_ant = get_model_trends(mat_sic_ant)

fig, (axtl, axtr) = plt.subplots(2, 1, sharex=True, figsize=(8,8))
fig.subplots_adjust(right=0.5)

axtl.plot(np.arange(1,13), trends.T, 'g')
axtr.plot(np.arange(1,13), trends_ant.T, 'g')

axtl.set_title('Arctic')
axtr.set_title('Antarctic')
axtl.set_ylabel(r'($\times 10^6$ km$^2$/decade)')
axtr.set_xlabel('Month')
axtr.set_xlim([1, 12])
#fig.savefig('monthly_sie_trends_kelly.pdf', bbox_inches='tight')
```
