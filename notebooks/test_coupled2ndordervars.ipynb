```python
# calc the vertical integral of poleward energy transport across 
#    a latitude (or all?) in units of Watts
# = 1/g int_pbot_to_ptop[  int_latcircle[ E*v a coslat dlon dp]]. a is Earth's radius

sea='ANN'
# 1. sum around lat circle. mult by a and coslat
# 2. sum vertically
# 3. divide by gravity

erad = con.get_earthrad()

across=65 # across which lat circle
latidx=cutl.find_nearest(lat,across)
latrad = np.deg2rad(lat[latidx])

fdict={'MPEF': basepath + 'kt131f_meridional_potential_energy_flux_2921-2931_ts.nc',
       'MSHF': basepath + 'kt131f_meridional_sens_heat_flux_2921-2931_ts.nc'}

field='MLT'
#fname=basepath + 'kt131f_meridional_potential_energy_flux_2921-2931_ts.nc'; conv=1 # z is already mult by gravity
#fname=basepath + 'kt131f_meridional_sens_heat_flux_2921-2931_ts.nc'; conv=Cp 
#fname=basepath + 'kt131f_meridional_latent_heat_flux_2921-2931_ts.nc'; conv=Lv
fname=basepath + 'kt131f_meridional_flux_zonal_mom_2921-2931_ts.nc'; conv=0.5 

#     mer SH flux: -8.5475296981e+13
#     mer PE flux: -3.67666789463e+14
#     mer LH flux: 8.04264121624e+12
#     mer ZonMom flux: 3.72973968486e+12

fld=cnc.getNCvar(fname,field,timesel='2922-01-01,2931-12-31')*conv

fldsub = np.squeeze(fld[:,:,latidx,:])
fldtm = np.mean(cutl.seasonalize_monthlyts(fldsub,season=sea),axis=0)
print fldtm.shape

fldint = np.sum(np.sum(fldtm,axis=1),axis=0)*erad*np.cos(latrad)

print fldint.shape
print fldint
```
