```python
# Add seasonal cycle of all obs/BCs here! @@
#import numpy as np
#cutl=reload(cutl)

field='sic'
debugsic=False

casenamecs = ('kemhadctl','kemnsidcctl','kemctl1') # add ens sims, etc?
casenameps = ('kemhadpert','kemnsidcpert','kem1pert2') # add ens sims, etc?
shortnames = ('Had','NSIDC','CanESM2')

figsuff= '' #'_50clim' # figure name suffix
latlim=70 

#print 'SEA CYCLE IS SHIFTED BY A YEAR@@ not anymore 7/28/14'
lat = cnc.getNCvar(con.getBCfilenames(field,casenamecs[0]),'lat')
lon = cnc.getNCvar(con.getBCfilenames(field,casenamecs[0]),'lon')

plotflds = np.zeros((3,12))
fldcs = np.zeros((3,12))
fldps = np.zeros((3,12))
for sii,casenamec in enumerate(casenamecs):
    
    casenamep = casenameps[sii]
    fldc = cnc.getNCvar(con.getBCfilenames(field,casenamec),field.upper())
    fldp = cnc.getNCvar(con.getBCfilenames(field,casenamep),field.upper())
    
    print fldc.shape
    fldc = fldc[1:,...] # get rid of first index, which is month 12 in year 0 (?)
    fldp = fldp[1:,...]
    fldc,fldcstd=cutl.climatologize(fldc) 
    fldp,fldpstd=cutl.climatologize(fldp)

    if field=='sicn':
        figsuff='totSIA'
        legloc='lower left'
        # convert to area
        fldc = cutl.calc_seaicearea(fldc,lat,lon)
        fldp = cutl.calc_seaicearea(fldp,lat,lon)
        
        #calc total area instead of average
        fldc = np.sum(np.sum(fldc[:,lat>0,:],axis=2),axis=1)
        fldp = np.sum(np.sum(fldp[:,lat>0,:],axis=2),axis=1)
        flddiff = fldp-fldc
        
    elif field=='sic':
        figsuff='pol' + str(latlim) + 'N3'
        legloc='lower left'
        # mean thickness in part of grid cell covered in ice: SIC/(SICN*deni) <=> SIT/SICN
        # what is the right way to show mean thickness? 
        #    over entire area or just sea-ice covered area?
        #fldc = cutl.polar_mean_areawgted3d(fldc/deni,lat,lon,latlim=latlim)
        #fldp = cutl.polar_mean_areawgted3d(fldp/deni,lat,lon,latlim=latlim)   
        
        sicnc = cnc.getNCvar(con.getBCfilenames('sicn',casenamec),'SICN')
        sicnp = cnc.getNCvar(con.getBCfilenames('sicn',casenamep),'SICN')
        sicnc = sicnc[1:,...]
        sicnp = sicnp[1:,...]
        sicnc,sicncstd = cutl.climatologize(sicnc)
        sicnp,sicnpstd = cutl.climatologize(sicnp)
        sicnc = ma.masked_where(sicnc==0,sicnc) # avoid div by zero
        sicnp = ma.masked_where(sicnp==0,sicnp) # avoid div by zero 
        fldc = ma.masked_where(fldc==0,fldc) # avoid div by zero 
        fldp = ma.masked_where(fldp==0,fldp) # avoid div by zero 
        
        if sii==1:
            fldcn=fldc
            fldpn=fldp
            # save nsidc spatial for debug
            tmpcn=fldc/(sicnc*deni)
            tmppn=fldp/(sicnp*deni)
        elif sii==0:
            fldch=fldc
            fldph=fldp
            # save hadisst spatial for debug
            tmpch=fldc/(sicnc*deni)
            tmpph=fldp/(sicnp*deni)
                
        #tmpc=fldc/(sicnc*deni)
        #tmpp=fldp/(sicnp*deni)
        tmpc=fldc/(deni)
        tmpp=fldp/(deni)
        
        flddiff = cutl.polar_mean_areawgted3d(tmpp-tmpc,lat,lon,latlim=latlim)
        fldc = cutl.polar_mean_areawgted3d(tmpc,lat,lon,latlim=latlim)
        fldp = cutl.polar_mean_areawgted3d(tmpp,lat,lon,latlim=latlim)   
        
    else: # gt!
        print '@@ GT not done!'
        latlim=40
        figsuff='pol' + str(latlim) + 'N'
        legloc='upper left'
        
        # mask out land
        lmask = con.get_t63landmask()
        lmask=np.tile(lmask,(12,1,1))
        
        fldc=ma.masked_where(lmask==-1,fldc)
        fldp=ma.masked_where(lmask==-1,fldp)
                
        flddiff=fldp-fldc
        
        sicnc = cnc.getNCvar(con.getBCfilenames('sicn',casenamec),'SICN')
        sicnp = cnc.getNCvar(con.getBCfilenames('sicn',casenamep),'SICN')
        sicnc = sicnc[1:,...]
        sicnp = sicnp[1:,...]
        sicnc,sicncstd = cutl.climatologize(sicnc)
        sicnp,sicnpstd = cutl.climatologize(sicnp)
        sicndiff=sicnp-sicnc
        
        # for climo: mask out where there the model considers it to be sea ice
        fldc=ma.masked_where(sicnc>=.15,fldc)
        fldp=ma.masked_where(sicnp>=.15,fldp)
        
        #flddiff=fldp-fldc
        
        # for the difference:
        # only include grid cells where sea ice changed
        
        # mask where control sea ice is zero
        flddiff=ma.masked_where(sicnc==0,flddiff)        
        # now mask where sicn didn't change
        #flddiff=ma.masked_where(sicndiff>-.10,flddiff)
        
        flddiff = cutl.polar_mean_areawgted3d(flddiff,lat,lon,latlim=latlim)
        fldc = cutl.polar_mean_areawgted3d(fldc,lat,lon,latlim=latlim)
        fldp = cutl.polar_mean_areawgted3d(fldp,lat,lon,latlim=latlim)
    
    fldcs[sii,:] = fldc #cutl.climatologize(fldc)  
    fldps[sii,:] = fldp # cutl.climatologize(fldp)  
    plotflds[sii,:] = flddiff #cutl.climatologize(fldp-fldc)
    
    #print plotfld.shape
    
```
